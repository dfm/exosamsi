```python
import numpy as np
import pyfits
```

## Load the data

```python
f = pyfits.open("kplr007902815-2010355172524_llc.fits")
data = f[1].data
mat = np.array(f[1].data, dtype=float)
f.close()
```

## Plot the raw SAP flux

```python
figure(figsize=(18, 6))
plot(data["time"], data["sap_flux"], ".k")
```

## Concatenate the auxiliary data

```python
A = np.vstack([data[k] for k in ["SAP_BKG", "POS_CORR1", "POS_CORR2"]]).T
b = data["sap_flux"]
x = data["time"]

mask = (np.isnan(b) + np.sum(np.isnan(A), axis=1)) == False
A = A[mask, :]
b = b[mask]
x = x[mask]

b -= np.median(b)
A -= np.median(A, axis=0)

A = np.hstack([A, np.ones((len(A), 1))])
A /= np.sqrt(np.sum(A ** 2, axis=0))
b /= np.sqrt(np.sum(b ** 2))
```

```python
figure(figsize=(18, 6))
plot(x, A, ".")
plot(x, b, ".k")
```

## Run linear least squares

```python
a, r, rank, s = np.linalg.lstsq(A, b)
a, r, rank, s
```

```python
figure(figsize=(18, 6))
plot(data["time"][mask], b - np.dot(A, a), ".k")
```

```python
figure(figsize=(18, 6))
plot(data["time"], data["sap_flux"], ".k")
```

## Higher order model

```python
import itertools
cols = ["SAP_BKG", "POS_CORR1", "POS_CORR2"]
A = np.vstack([data[k] for k in cols] +
              [data[k1] * data[k2]
               for k1, k2 in itertools.product(cols, cols)]).T

A = A[mask, :]
A -= np.median(A, axis=0)
A = np.hstack([A, np.ones((len(A), 1))])
A /= np.sqrt(np.sum(A ** 2, axis=0))
```

```python
figure(figsize=(18, 6))
plot(x, A, ".")
plot(x, b, ".k")
```

## Run least squares again

```python
a, r, rank, s = np.linalg.lstsq(A, b)
a, r, rank, s
```

```python
figure(figsize=(18, 6))
plot(data["time"][mask], b - np.dot(A, a), ".k")
```
