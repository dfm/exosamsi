```python
import numpy as np
import pyfits
```

```python
f = pyfits.open("kplr010395543-2010355172524_llc.fits")
data = f[1].data
mat = np.array(f[1].data, dtype=float)
f.close()
```

```python
plot(data["time"], data["sap_flux"], ".k")
```

```python
A = np.vstack([data[k] for k in ["SAP_BKG", "POS_CORR1", "POS_CORR2"]]).T
b = data["sap_flux"]
x = data["time"]

mask = ~np.isnan(b)
A = A[mask, :]
b = b[mask]
x = x[mask]

b -= np.median(b)

A -= np.median(A, axis=0)
```

```python
plot(x, A[:, 1], ".")
# plot(x, b, ".k")
```

```python
a, r, rank, s = np.linalg.lstsq(A, b)
a, r, rank, s
```

```python
figure(figsize=(18, 6))
plot(data["time"][mask], b - np.dot(A, a), ".")
```
